### Creación de Series a partir de un array de NumPy

```python
# Creo un arreglo de numeritos al azar usando la librería numpy
import numpy as np
import pandas as pd

arr_aleatorio = np.random.randint(0, 100, size=10)
serie_aleatoria = pd.Series(arr_aleatorio)
print("Serie aleatoria:")
print(serie_aleatoria)
```

```python
# Hago una lista rápida del 1 al 10 sin tener que escribirlos uno por uno
arr_rango = np.arange(1, 11)
serie_rango = pd.Series(arr_rango)
print("Serie del 1 al 10:")
print(serie_rango)
```

```python
# Pongo números con decimales en mi arreglo para luego hacer mi serie
arr_decimal = np.array([1.1, 2.2, 3.3, 4.4])
serie_decimal = pd.Series(arr_decimal)
print("Serie de decimales:")
print(serie_decimal)
```

```python
# Transformo todas las temperaturas de golpe aplicando una fórmula matemática simple
arr_celsius = np.array([0, 20, 25, 30, 40, 100])
serie_celsius = pd.Series(arr_celsius)
serie_fahrenheit = (serie_celsius * 9/5) + 32
print("Temperaturas en Fahrenheit:")
print(serie_fahrenheit)
```

```python
# Multiplico todos los elementos por 10 al mismo tiempo sin usar ciclos
arr_5 = np.array([1, 2, 3, 4, 5])
serie_multiplicada = pd.Series(arr_5) * 10
print("Serie de array multiplicada por 10:")
print(serie_multiplicada)
```

### Especificación de un índice personalizado

## 7️⃣ Series con datos de diferentes tipos

```python
# ¡Sorpresa! Pandas me deja mezclar números, texto y cajas vacías en el mismo lugar
import pandas as pd
import numpy as np
serie_mixta_1 = pd.Series([10, 'Hola', np.nan, 42])
print(serie_mixta_1)
```

```python
# Le pregunto a la serie de qué tipo de datos está compuesta por dentro exactamente
print("dtype:", serie_mixta_1.dtype)
```

```python
# Al mezclar cosas tan distintas, todo se guarda de forma super general bajo un solo título ('Object')
serie_mixta_2 = pd.Series([1, 'dos', 3.0, True])
print(serie_mixta_2)
```

```python
# Guardo hasta fechas junto con texto y números, se forma una ensalada de datos interesante
serie_mixta_3 = pd.Series(['2023-01-01', 100, 'texto', pd.to_datetime('2023-01-02')])
print(serie_mixta_3)
```

```python
# Obligo a que absolutamente todos los elementos se vuelvan cadenas de texto plano
serie_str = serie_mixta_2.astype(str)
print(serie_str)
print(serie_str.dtype)
```

## 🔟 Operaciones y manipulación sobre Series

### 10.1 Operaciones aritméticas

```python
# Hago una suma elemento por elemento entre mis dos grupitos de números
s1 = pd.Series([10, 20, 30])
s2 = pd.Series([5, 10, 15])
print(s1 + s2)
```

```python
# Le quito o resto los números del segundo grupo a los del primero en línea
print(s1 - s2)
```

```python
# Agarro cada número del grupo y lo inflo multiplicándolo por el doble al instante
print(s1 * 2)
```

```python
# Le saco la mitad de forma directa a todos mis valores sin pensarlo
print(s1 / 2)
```

```python
# Hago una suma elemento por elemento entre mis dos grupitos de números
print((s1 + s2) / 2)
```

### 10.2 Aplicación de funciones: `apply` y `map`

```python
# Le aplico una regla invisible o anónima a cada valor para potenciarlo o elevarlo al cuadrado
s_num = pd.Series([1, 2, 3, 4, 5])
print(s_num.apply(lambda x: x**2))
```

```python
# Obligo a que todos los nombres se pongan en letras grandotas (mayúsculas) para que resalten
s_nombres = pd.Series(['juan', 'ana', 'luis'])
print(s_nombres.map(str.upper))
```

```python
# Le paso la herramienta mágica de la raíz cuadrada a cada número para que la ejecute uno a uno
import numpy as np
print(s_num.apply(np.sqrt))
```

```python
# Me invento mi propia regla clasificador y se la paso al grupo con ayuda del apply
def clasificar(val):
    return 'Alto' if val > 3 else 'Bajo'
print(s_num.apply(clasificar))
```

```python
# Traigo mi propio diccionario de inglés y uso map para traducir todo de un solo jalón
s_colores = pd.Series(['rojo', 'azul', 'verde'])
dicc = {'rojo': 'red', 'azul': 'blue', 'verde': 'green'}
print(s_colores.map(dicc))
```

### 10.3 Ordenación y clasificación

### 10.4 Filtrado y selección condicional

```python
# Creo un colador o filtro estricto para dejar por fuera a todos los menores o iguales de 50
s_filtro = pd.Series([10, 40, 60, 80, 100])
print(s_filtro[s_filtro > 50])
```

```python
# Me saco de manga una táctica para quedarme únicamente con aquellos que sean totalmente pares
print(s_filtro[s_filtro % 2 == 0])
```

```python
# Busco minuciosamente entre las frases de texto solo la que incluya la palabrita clave 'hola'
s_frases = pd.Series(['hola mundo', 'adios mundo', 'hola alien'])
print(s_frases[s_frases.str.contains('hola')])
```

```python
# Me pongo duro y exijo dos reglas a la vez: que sean mayores a 30 'Y' que no pasen de 70
print(s_filtro[(s_filtro > 30) & (s_filtro < 70)])
```

```python
# Filtro los datos positivos y pido una lista con sus identificadores de una vez
valida = s_filtro[s_filtro > 0]
print(valida.index.tolist())
```

### 10.5 Manejo de valores nulos

```python
# Relleno todos esos huecos fastidiosos que no tienen dato usando un cero de comodín
s_nula = pd.Series([1, np.nan, 3, np.nan, 5])
print(s_nula.fillna(0))
```

```python
# Soy drástico y boto a la papelera aquellas filas que me vengan con el cuento de un valor nulo
print(s_nula.dropna())
```

```python
# Me pongo los lentes y cuento con mis dedos cuántos huecos o faltantes me salieron al final
print(s_nula.isna().sum())
```

```python
# Para no arruinar mis matemáticas metiendo ceros, mejor disimulo los vacíos usando un promedio
print(s_nula.fillna(s_nula.mean()))
```

```python
# Pongo a la fuerza un número rarísimo y ridículo como el -1 para poder notar rapidito dónde me faltó un dato
print(s_nula.fillna(-1))
```

### 10.6 Operaciones de alineación y combinación

```python
# Sumo dos listas distintas pero Pandas iguala los nombres que se parecen y donde no hay coincidencia tira un nulo
sa = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
sb = pd.Series([4, 5, 6], index=['b', 'c', 'd'])
print(sa + sb)
```

```python
# Para que la suma no tenga errores ni me bote vacíos, primero coloco ceros en el fondo donde falte algo y ahí sí sumo
print(sa.add(sb, fill_value=0))
```

```python
# Si mi lista A principal falla y tiene un vacío, le digo que se asome a la lista B y se copie el dato de ahí
print(sa.combine_first(sb))
```

```python
# Resto dos series que medio se cruzan un poquito y compruebo el caos total quedando errores en los bordes sueltos
sc = pd.Series([10, 20], index=[1, 2])
sd = pd.Series([30, 40], index=[2, 3])
print(sc - sd)
```

```python
# Hago tooda una mezcla multiplicando y al rato sumando pero previendo los valores rotos o perdidos usando relleno
print((sa * 2).add(sb, fill_value=1))
```

### 10.7 Reindexación

```python
# Engrando mi maleta de etiquetas trayendo un cupo nuevo ('e'), aunque por ahorita le toque venir sin nada adentro
s_reidx = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(s_reidx.reindex(['a', 'b', 'c', 'd', 'e']))
```

```python
# Pido que las etiquetas de nombres se barajeen y empujen el orden a que todo me quede literalmente de cabeza
s_num_idx = pd.Series([10, 20, 30], index=[1, 2, 3])
print(s_num_idx.reindex([3, 2, 1]))
```

```python
# Agrando de nuevo mi seriecita pero esta vez me cercioro de meterles unos ceros de cortesía a los huéspedes recién traídos
print(s_reidx.reindex(['a', 'b', 'c', 'd', 'e'], fill_value=0))
```

```python
# En vez de engañarlos con ceros, relleno a los huecos con el promedio solidario de lo que ya estaba de antes
nuevo_indice = ['a', 'b', 'c', 'd', 'e']
print(s_reidx.reindex(nuevo_indice).fillna(s_reidx.mean()))
```

```python
# Elimino algunos índices viejos del pasado y agrego índices fresquitos, revolviendo o reajustando la organización
print(s_reidx.reindex(['b', 'c', 'e', 'f']))
```
